* 49 Group Anagram

```python
# hashmap 
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        res = defaultdict(list)

        for s in strs:
            char = [0] * 26

            for c in s:
                char[ord(c) - ord('a')] += 1
            
            res[tuple(char)].append(s)
        
        return list(res.values())
```
