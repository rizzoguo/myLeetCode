* 347 Top K Frequent Elements

```python
# Bucket Sort
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
      count = {}
      freq = [[] for _ in range(len(nums) + 1)]
      res = []
      for num in nums:
        count[num] = 1 + count.get(num, 0)
      for num, cnt in count.items():
        freq[cnt].append(num)
      
      for i in range(len(freq) - 1, 0, -1):
        for num in freq[i]:
          res.append(num)
          if len(res) == k:
            return res
```
